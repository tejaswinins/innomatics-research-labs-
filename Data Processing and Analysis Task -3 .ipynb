Problem Statement 1: Employee Performance Bonus Eligibility

```python
def employee_bonus_eligibility(employees):
    """
    Identifies top performers eligible for bonus.
    Handles tie cases automatically.
    """
    # Find highest performance score
    highest_score = max(employees.values())
    # Find all employees with highest score
    top_performers = [name for name, score in employees.items() if score == highest_score]

    print("Top Performers Eligible for Bonus:", 
          ", ".join(top_performers), 
          f"(Score: {highest_score})")
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
employee_bonus_eligibility(employees)
```

Problem Statement 2: Search Query Keyword Analysis

```python
import string
def keyword_analysis(query):
    """
    Converts query to lowercase,
    removes punctuation,
    counts frequency,
    displays words searched more than once.
    """
    # Convert to lowercase
    query = query.lower()
    # Remove punctuation
    query = query.translate(str.maketrans("", "", string.punctuation))
    words = query.split()
    frequency = {}
    # Count word frequency
    for word in words:
        frequency[word] = frequency.get(word, 0) + 1
    # Filter words with frequency > 1
    result = {word: count for word, count in frequency.items() if count > 1}
    print(result)
keyword_analysis("Buy mobile phone buy phone online")
```

Problem Statement 3: Sensor Data Validation

```python
def validate_sensor_data(sensor_readings):
    """
    Identifies valid (even) readings.
    Stores as (hour_index, reading_value) pairs.
    """
    valid_readings = []
    for index, value in enumerate(sensor_readings):
        if value % 2 == 0:  # Even number check
            valid_readings.append((index, value))
    print("Valid Sensor Readings (Hour, Value):")
    print(valid_readings)
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
validate_sensor_data(sensor_readings)
```

Problem Statement 4: Email Domain Usage Analysis

```python
def email_domain_analysis(emails):
    """
    Counts domain usage
    Calculates percentage usage
    """
    domain_count = {}
    for email in emails:
        # Extract domain after '@'
        domain = email.split("@")[1]
        domain_count[domain] = domain_count.get(domain, 0) + 1
    total_emails = len(emails)
    # Display percentage usage
    for domain, count in domain_count.items():
        percentage = (count / total_emails) * 100
        print(f"{domain}: {percentage:.0f}%")
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
email_domain_analysis(emails)
```

Problem Statement 6: Duplicate User ID Detection

```python
# Input list of user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
# Dictionary to store frequency of each user ID
frequency = {}
# Count occurrences of each user ID
for user in user_ids:
    if user in frequency:
        frequency[user] += 1
    else:
        frequency[user] = 1
# Display only duplicate user IDs (count > 1)
for user, count in frequency.items():
    if count > 1:
        print(f"{user} → {count} times")
```
