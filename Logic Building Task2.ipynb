Problem 1: Unique Words in a Sentence

```python
sentence = input("Enter sentence: ")
words = sentence.split()
unique_words = []
seen = set()
for word in words:
    if word not in seen:
        unique_words.append(word)
        seen.add(word)
print("\nUnique words count:", len(unique_words))
print("Unique words:", unique_words)
```

Problem 2: Highest Salary from Employee Data

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]
print("Highest Salary:", highest_employee, "-", highest_salary)
```

Problem 3: Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]
maximum = numbers[0]
minimum = numbers[0]
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

Problem 4: Count Products Above a Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]
count = 0
for price in prices:
    if price > 1000:
        count += 1
print("Products above 1000:", count)
```

Problem 5: Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]
total_days = len(attendance)
present_days = 0
for status in attendance:
    if status == "P":
        present_days += 1
attendance_percentage = (present_days / total_days) * 100
print("Attendance Percentage:", attendance_percentage)
```

Problem 7: Count Character Frequency

```python
text = "pythonp"
char_frequency = {}
for char in text:
    if char in char_frequency:
        char_frequency[char] += 1 
    else:
        char_frequency[char] = 1  
print(char_frequency)
```

Problem 10: Calculate Average Marks

```python
marks = [65, 70, 80, 75]
total_marks = 0
for mark in marks:
    total_marks += mark
number_of_subjects = len(marks)
average = total_marks / number_of_subjects
print("Average Marks:", average)
```
